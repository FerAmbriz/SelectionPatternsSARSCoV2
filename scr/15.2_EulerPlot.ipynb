```r
df <- read_tsv('../../Data/Estratos/miRNA_predichos_ConstantClustersEuler.tsv')
df
```

```r
G1 <- na.omit(df$miRNA_Estrato_0)
G2 <-na.omit(df$miRNA_Estrato_1)

#myV <- plotVenn(list(G1=G1, G2=G2), outFile = 'a.svg', nCycles = 20000)
```

```r
ggVennDiagram(list(G1=G1, G2=G2)) +
 theme(legend.position = "none") +
 scale_fill_gradient(low = "#aed6f1", high = "#73c6b6")
```

```r
#pdf('../../Plots/VennConstantClusters.pdf')
pdf(file = "../../Plots/VennConstantClusters.pdf")


# Define tus conjuntos (G1 y G2)
conjuntos <- list(G1 = G1, G2 = G2)


venn.plot <- venn.diagram(
  x = conjuntos,
  filename = NULL,
  fill = c("#76d7c4", "#ff922c"),  # Colores de relleno para los círculos
  col = "white",  # Color de la línea del círculo (puedes cambiarlo a tu preferencia)
  cex = 2,  # Tamaño de la letra
  lwd = 2,  # Ancho de los bordes
  cat.cex = 3
)


grid.draw(venn.plot)
dev.off()
```
