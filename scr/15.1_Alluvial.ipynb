```r
library(ggplot2)
library(ggalluvial)
library(tidyverse)
library(ggsankey)
```

```r
df = read_tsv('../../Data/Alluvial.tsv')
df
```

```r
options(repr.plot.width = 15, repr.plot.height = 10)

#pdf('../../Plots/AlluvialPlot.pdf', width = 17, height = 10)

ggplot(df, aes(x = tiempo, stratum = Via, alluvium = Via, y = RF, fill = Via)) +
  geom_flow(stat = "alluvium", lode.guidance = "frontback") +
  geom_stratum(color = NA) +
  scale_x_discrete(limits = c("(2010-12-05,2020-03-01]","(2020-03-01,2020-09-01]", "(2020-09-01,2021-01-01]", "(2021-01-01,2021-05-01]",
  "(2021-05-01,2022-01-01]","(2022-01-01,2022-06-01]","(2022-06-01,2022-11-01]", "(2022-11-01,2023-01-01]","(2023-01-01,2023-07-01]",
  "(2023-07-01,2023-09-06]"), expand = c(0.01, 0.05)) +
  theme_minimal() +
  theme(legend.position = "bottom",
        axis.text.x = element_text(angle = 90, hjust = 1)) +
  labs(title = "Vias metabolicas reguladas a través de t", x = "Tiempo", y = "RF")
#dev.off()
```

```r
df_sankey <- df %>%
  make_long(Via, tiempo)
df_sankey
```

```r
ggplot(df_sankey, aes(x = x, next_x = next_x, node = node, next_node = next_node, fill = factor(node), label = node)) +
  geom_sankey(flow.alpha = 0.5, node.color = "black") +
  geom_sankey_label(size = 3, color = "black", fill = "white") +
  theme_sankey(base_size = 16) +
  labs(title = "Gráfico de Sankey de la Regulación Positiva de Proliferación Celular")
```
