# Entropy

```python
df5 = df_merge[df_merge['Region']== "5'UTR"]
df3 = df_merge[df_merge['Region']== "3'UTR"]
```

```python
#Shapiro-Wilk
stat, p_value = stats.shapiro(df5['H'])
print(f'EstadÃ­stico: {stat}, p-valor: {p_value}')

# Agrupar los datos por la variable categÃ³rica
groups = [group['H'].values for name, group in df5.groupby('Type')]

# Realizar la prueba de Kruskal-Wallis
kruskal_result = stats.kruskal(*groups)
print(kruskal_result)

dunn_result = sp.posthoc_dunn(df5, val_col='H', group_col='Type')
print(dunn_result)
```

```python
#Shapiro-Wilk
stat, p_value = stats.shapiro(df3['H'])
print(f'EstadÃ­stico: {stat}, p-valor: {p_value}')

# Agrupar los datos por la variable categÃ³rica
groups = [group['H'].values for name, group in df3.groupby('Type')]

# Realizar la prueba de Kruskal-Wallis
kruskal_result = stats.kruskal(*groups)
print(kruskal_result)

dunn_result = sp.posthoc_dunn(df3, val_col='H', group_col='Type')
print(dunn_result)
```
