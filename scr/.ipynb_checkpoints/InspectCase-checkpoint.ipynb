```python
df = pd.DataFrame(df_lc21j['country'].value_counts()).reset_index()
d = {
    'Bosnia and Herzegovina': 'Bosnia and Herz.',
    'Czech Republic': 'Czechia',
    'Dominican Republic': 'Dominican Rep.',
    'USA': 'United States of America'
}

df['country'] = df['country'].replace(d)
df
```

```python
shapefile_path = "../../Data/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp"

# Read the shapefile
world = gpd.read_file(shapefile_path)
world.head()
```

```python
lst = []
for i in world['SOVEREIGNT']:
    if i not in list(df['country']):
        lst.append(0)
    else:
        lst.append(list(df[df['country'] == i]['count'])[0])
world['Cluster'] = lst
world.head()
```

```python
world = world[(world['SOVEREIGNT']=='United States of America') | (world['SOVEREIGNT']=='Mexico') | (world['SOVEREIGNT']=='Canada')]
world
```

```python
fig, ax = plt.subplots(1, 1, figsize = (30, 5))

# Crear un diccionario de colores
color_dict = {0: '#34495E', 1: '#45B39D', 2: '#F7DC6F', 3: '#61A1EC', 4: '#EC8361', 12180:'#82e0aa', 10172: '#5dade2', 12177: '#82e0aa'}

# Crear una lista de colores para cada valor en 'Cluster'
colors = [color_dict[val] for val in world['Cluster']]

# Dibujar el gráfico con los colores especificados
world.plot(column='Cluster', color=colors, ax=ax)

ax.set_axis_off()
```

```python
df_e = pd.read_csv('/home/ferambriz/Projects/SARS-CoV-2/Data/test_usa_mexico/headers/entropy.tsv', sep = '\t')
df_e
```

```python
DF = pd.DataFrame({'Site': [8782, 14408, 14408, 23403, 21618, 26767, 28461, 11332, 19220], 'Variant':['21J'] * 9, 'Certainty': [1]*9})

# Lista de todos los sitios posibles
all_sites = range(1, max(DF['Site']) + 1)

# Crear un DataFrame con todos los sitios y certeza 0
new_data = {
    'Site': [site for site in all_sites if site not in DF['Site'].values],
    'Variant': ['21J'] * len([site for site in all_sites if site not in DF['Site'].values]),
    'Certainty': [0] * len([site for site in all_sites if site not in DF['Site'].values])
}
new_df = pd.DataFrame(new_data)

# Combinar los DataFrames
DF = pd.concat([DF, new_df]).sort_values(by='Site').reset_index(drop=True)
DF
```

```python
UTR5 = df_e[df_e['Region'].isin(["5'UTR"])]
UTR3 = df_e[df_e['Region'].isin(["3'UTR"])]
```

```python
sns.lineplot(data=UTR5, x="Site", y='H', hue='Region', legend=False, palette = 'Set2')
sns.despine(left=True, bottom=True)
plt.savefig('../../Plots/5UTR_inspectCase.pdf')
```

```python
sns.lineplot(data=UTR3, x="Site", y='H', hue='Region', legend=False)
sns.despine(left=True, bottom=True)
plt.savefig('../../Plots/3UTR_inspectCase.pdf')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear la figura y los subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(30, 10), sharex=True)

# Primer gráfico
sns.lineplot(data=df_e, x="Site", y='H', hue='Region', legend=False, ax=ax1, palette='Set2', alpha=0.4)
ax1.set_title('F', fontweight='bold', loc='left')
ax1.set_title("Entropy", fontsize=10, pad=20)
ax1.set_ylabel('Entropy')

# Segundo gráfico
sns.lineplot(data=DF, x="Site", y='Certainty', hue='Variant', legend=False, ax=ax2, palette=['red'])
ax2.set_ylabel('Certainty')
ax2.invert_yaxis()  # Invertir el eje y

# Ajustar el diseño
plt.xticks(rotation=45)
sns.despine(left=True, bottom=True)
plt.tight_layout()

# Ajustar la separación entre subplots
plt.subplots_adjust(hspace=-0.1, top=0, bottom=0.2)  

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# Crear una figura con GridSpec
fig = plt.figure(figsize=(30, 12))
gs = GridSpec(2, 4, figure=fig)

# Primer gráfico
ax1 = fig.add_subplot(gs[0, 0])
color_dict = {0: '#34495E', 1: '#45B39D', 2: '#F7DC6F', 3: '#61A1EC', 4: '#EC8361', 12180:'#82e0aa', 10172: '#5dade2', 12177: '#82e0aa'}
colors = [color_dict[val] for val in world['Cluster']]
world.plot(column='Cluster', color=colors, ax=ax1)
ax1.set_axis_off()

# Segundo gráfico
ax2 = fig.add_subplot(gs[0, 1:4])
sns.lineplot(data=DF, x="Site", y='Certainty', hue='Variant', legend=False, ax=ax2, palette=['red'], linestyle='--')
ax2.set_ylabel('H')

# Segundo gráfico con un segundo eje y
ax3 = ax2.twinx()
sns.lineplot(data=df_e, x="Site", y='H', hue='Region', ax=ax3, palette='Set2')
ax3.set_ylabel('Marcador')

# Ajustar el diseño
plt.xticks(rotation=45)
sns.despine(left=True, bottom=True)
plt.tight_layout()

handles, labels = ax3.get_legend_handles_labels()
ax3.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.1, 1), ncol=2)

# Tercer gráfico
ax4 = fig.add_subplot(gs[1, 0:2])
sns.lineplot(data=UTR5, x="Site", y='H', hue='Region', legend=False, palette='Set2', ax=ax4)
sns.despine(left=True, bottom=True)

# Cuarto gráfico
ax5 = fig.add_subplot(gs[1, 2:4])
sns.lineplot(data=UTR3, x="Site", y='H', hue='Region', legend=False, palette=['gray'], ax=ax5)
sns.despine(left=True, bottom=True)


ax1.set_title('A', fontweight="bold", loc = 'left', fontsize = 20)
ax3.set_title('B', fontweight="bold", loc = 'left', fontsize = 20)
ax4.set_title('C', fontweight="bold", loc = 'left', fontsize = 20)
ax5.set_title('D', fontweight="bold", loc = 'left', fontsize = 20)

ax2.set_xlabel('Sitio') 
ax3.set_xlabel('Sitio')  # Ajusta el valor de fontsize según sea necesario
ax4.set_xlabel('Sitio') 
ax5.set_xlabel('Sitio') 

plt.tight_layout()


plt.savefig('../../Plots/USA-MEXICO_21J.pdf')
plt.show()
```
