```python
import pandas as pd
import os
import re
import ast
```

# Extraccion de secuencias conservadas

```python
files = os.listdir('../Data/concensus_seq/')

expr = r".*\.fasta$"
files = [cadena for cadena in files if re.search(expr, cadena)]
files[0:3]
```

```python
def extract_seq(df):
    lst_seq = []
    for i,j in zip(df.Variante, df.RegionesConservadas):
        
        for filename in files:
            #Relacionar ID con archivo
            if filename == 'concensus.fasta':
                ID = 'Global'
                file =  filename
            else:
                ID = filename.replace('concensus_sequence_', '').replace('.fasta', '')
                file =  filename

            # Cargar e identificar la secuencia concenso
            if ID == i: 
                 with open('../Data/concensus_seq/' + filename) as file:
                     content = file.read()
                     lst_j = ast.literal_eval(j)
                     lst_j.sort()
                     
                     seq = content[lst_j[0]-1:lst_j[len(lst_j)-1]]
                     lst_seq.append(seq)
    return lst_seq
```

```python
def preprocessing(df):
    lst_headers = []
    for i, j in zip(df.RegionesConservadas, df.Variante): 
        x = i.replace('[', '').replace(']', '').split(', ')
        y = '>'+ j + ':' +x[0] + '-' + x[len(x)-1]
        lst_headers.append(y)
    df['Headers'] = lst_headers
    return df
```

```python
df_t = pd.read_csv('../Data/RegionesConsecutivasConservadasTimes.tsv', sep = '\t')
df_t = preprocessing(df_t)

df_t['Seq'] = extract_seq(df_t)
df_t
```

```python
df_v = pd.read_csv('../Data/RegionesConsecutivasConservadasSinFoldVariants.tsv', sep = '\t')
df_v = preprocessing(df_v)

df_v['Seq'] = extract_seq(df_v)
df_v
```

# GeneraciÃ³n de FASTA

```python
with open('../Data/FastaPredict/conserved_regions_UTR_time.fasta', 'w') as f:
    for header, seq in zip(df_t.Headers, df_t.Seq):
        f.write(header + '\n')
        f.write(seq + '\n')
        
with open('../Data/FastaPredict/conserved_SinFold_regions_UTR_variants.fasta', 'w') as f:
    for header, seq in zip(df_v.Headers, df_v.Seq):
        f.write(header + '\n')
        f.write(seq + '\n')
```
