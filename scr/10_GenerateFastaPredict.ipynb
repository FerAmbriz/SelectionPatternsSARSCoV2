```python
import pandas as pd
import os
import re
import ast
```

# Extraccion de secuencias conservadas

```python
path = '../../Data/Estratos/concensus_seq/'
folder = os.listdir(path)
folder.pop(0)
```

```python
renamed_files = {'Gst1':'cluster_0_gt', 'Gst2':'cluster_1_gt', 
 'G1':'cluster_0', 'G2':'cluster_1' ,'G3':'cluster_2' , 'G4':'cluster_3', 'G5': 'cluster_4',
 'Gs1': 'cluster_0_sp', 'Gs2': 'cluster_1_sp', 'Gs3':'cluster_2_sp' , 'Gs4':'cluster_3_sp'
}
to_rename = list(renamed_files.keys())
```

```python
def preprocessing(df):
    lst_headers = []
    for i, j in zip(df.RegionesConservadas, df.Variante): 
        x = i.replace('[', '').replace(']', '').split(', ')
        y = '>'+ j + ':' +x[0] + '-' + x[len(x)-1]
        lst_headers.append(y)
    df['Headers'] = lst_headers
    return df
    
def extract_seq(df):
    lst_seq = []
    df = df[df['Variante']!= 'Global']
    for i in df.Variante.unique(): 
        df_i = df[df['Variante']==i]
        
        for k in folder:
            files = os.listdir(f'{path}/{k}/')
            expr = r".*\.fasta$"
            files = [cadena for cadena in files if re.search(expr, cadena)]


            
            for filename in files:
                if i in to_rename:
                    ii = renamed_files[i]; ID = filename.replace('concensus_', '').replace('.fasta', '')
                else:
                    ii = i; ID = filename.replace('concensus_sequence_', '').replace('.fasta', '')
                
                if ii == ID:
                    file = filename
                    with open(f'{path}/{k}/{filename}') as file:
                        content = file.read()
                        for j in df_i.RegionesConservadas:
                            lst_j = ast.literal_eval(j)
                            lst_j.sort()
                                 
                            seq = content[lst_j[0]-1:lst_j[len(lst_j)-1]]
                            lst_seq.append(seq)
    df = preprocessing(df)
    df['Seq'] = lst_seq
    return df
```

```python
df_t = pd.read_csv('../../Data/Estratos/consecutive_conserved/RegionesConsecutivasConservadasTimes.tsv', sep = '\t')
df_t = extract_seq(df_t)
df_t
```

```python
df_v = pd.read_csv('../../Data/Estratos/consecutive_conserved/RegionesConsecutivasConservadasSinFoldVariants.tsv', sep = '\t')
df_v = extract_seq(df_v)
df_v
```

```python
df_gt = pd.read_csv('../../Data/Estratos/consecutive_conserved/RegionesConsecutivasConservadasGrafosConstantes.tsv', sep = '\t')
df_gt = extract_seq(df_gt)
df_gt
```

```python
df_s = pd.read_csv('../../Data/Estratos/consecutive_conserved/RegionesConsecutivasConservadasSpectral.tsv', sep = '\t')
df_s = extract_seq(df_s)
df_s
```

```python
df_sp = pd.read_csv('../../Data/Estratos/consecutive_conserved/RegionesConsecutivasConservadasSpectrum.tsv', sep = '\t')
df_sp = extract_seq(df_sp)
df_sp
```

# GeneraciÃ³n de FASTA

```python
with open('../../Data/Estratos/FastaPredict/conserved_regions_UTR_time.fasta', 'w') as f:
    for header, seq in zip(df_t.Headers, df_t.Seq):
        f.write(header + '\n')
        f.write(seq + '\n')
        
with open('../../Data/Estratos/FastaPredict/conserved_SinFold_regions_UTR_variants.fasta', 'w') as f:
    for header, seq in zip(df_v.Headers, df_v.Seq):
        f.write(header + '\n')
        f.write(seq + '\n')

        
with open('../../Data/Estratos/FastaPredict/conserved_regions_UTR_ConstantClusters.fasta', 'w') as f:
    for header, seq in zip(df_gt.Headers, df_gt.Seq):
        f.write(header + '\n')
        f.write(seq + '\n')
        
with open('../../Data/Estratos/FastaPredict/conserved_regions_UTR_spectral.fasta', 'w') as f:
    for header, seq in zip(df_s.Headers, df_s.Seq):
        f.write(header + '\n')
        f.write(seq + '\n')
        
with open('../../Data/Estratos/FastaPredict/conserved_regions_UTR_spectrum.fasta', 'w') as f:
    for header, seq in zip(df_sp.Headers, df_sp.Seq):
        f.write(header + '\n')
        f.write(seq + '\n')
```
