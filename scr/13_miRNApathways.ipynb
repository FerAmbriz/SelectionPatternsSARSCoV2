# GO

* GO_BP_prediction_intersection.csv: Este archivo podría contener las predicciones de la función biológica (Biological Process, BP) de ciertos genes que se han obtenido a través de varios métodos y que coinciden entre sí (intersección).
* GO_BP_prediction_union.csv: Este archivo podría contener todas las predicciones de la función biológica de ciertos genes que se han obtenido a través de varios métodos, incluyendo aquellas que no coinciden entre sí (unión).
* GO_BP_validated_miRTarBase_all.csv: Este archivo podría contener las funciones biológicas de ciertos genes que han sido validadas y están presentes en la base de datos miRTarBase.
* GO_BP_validated_miRTarBase_strong.csv: Este archivo podría contener las funciones biológicas de ciertos genes que han sido validadas con evidencia fuerte y están presentes en la base de datos miRTarBase.
Los archivos con “CC” en su nombre se refieren a la localización celular (Cellular Component, CC) de los genes, y los archivos con “MF” en su nombre se refieren a la función molecular (Molecular Function, MF) de los genes.

```python
pathways = pd.read_csv('../../Data/DataBasePathway/hsa_p_value_matrices/GO_BP_validated_miRTarBase_strong.csv')
pathways.head()
```

```python
n = 16
miRNA_validated = list(pathways.index)
def process(params):
    i, j, k, l, m, o, p = params
    if i in miRNA_validated:
        df_i = pd.DataFrame(pathways.loc[i].T).reset_index()
        df_i.columns = ['function', 'p_value']
        df_i = df_i[df_i['p_value']<=0.05]
        df_i['miRNA'] = i; df_i['Variant'] = j; df_i['DuplexEnergy'] = k; df_i['Jaccard']=l; df_i['Region'] = m
        df_i['UTR_region'] = o; df_i['InteractionRegionUTR'] = p
        return df_i
    else:
        return pd.DataFrame()
```

```python
with Pool(n) as p:
    resultst = p.map(process, zip(dft['miRNA_lite'], dft['Variant'], dft['DuplexEnergy'], dft['Jaccard'], 
                                  dft['Region'], dft['UTR_region'], dft['InteractionRegionUTR']))

df_bdt = pd.concat(resultst)
df_bdt
```

```python
with Pool(n) as p:
    resultsv = p.map(process, zip(dfv['miRNA_lite'], dfv['Variant'], dfv['DuplexEnergy'], dfv['Jaccard'], 
                                  dfv['Region'], dfv['UTR_region'], dfv['InteractionRegionUTR']))

df_bdv = pd.concat(resultsv)
df_bdv
```

```python
with Pool(n) as p:
    resultss = p.map(process, zip(dfs['miRNA_lite'], dfs['Variant'], dfs['DuplexEnergy'], dfs['Jaccard'], 
                                  dfs['Region'], dfs['UTR_region'], dfs['InteractionRegionUTR']))

df_bds = pd.concat(resultss)
df_bds
```

```python
with Pool(n) as p:
    resultsp = p.map(process, zip(dfsp['miRNA_lite'], dfsp['Variant'], dfsp['DuplexEnergy'], dfsp['Jaccard'], 
                                  dfsp['Region'], dfsp['UTR_region'], dfsp['InteractionRegionUTR']))

df_bdsp = pd.concat(resultsp)
df_bdsp
```

```python
with Pool(n) as p:
    resultgt =  p.map(process, zip(dfgt['miRNA_lite'], dfgt['Variant'], dfgt['DuplexEnergy'], dfgt['Jaccard'], 
                                  dfgt['Region'], dfgt['UTR_region'], dfgt['InteractionRegionUTR']))

df_bdgt = pd.concat(resultgt)
df_bdgt
```

```python
df_bdt.to_csv('../../Data/Estratos/Pathways/Contribution_miRNA_time_pathwayGO_BP.tsv', sep = '\t', index = False)
df_bdv.to_csv('../../Data/Estratos/Pathways/Contribution_miRNA_variant_pathwayGO_BP.tsv', sep = '\t', index = False)
df_bds.to_csv('../../Data/Estratos/Pathways/Contribution_miRNA_spectral_pathwayGO_BP.tsv', sep = '\t', index = False)
df_bdsp.to_csv('../../Data/Estratos/Pathways/Contribution_miRNA_spectrum_pathwayGO_BP.tsv', sep = '\t', index = False)
df_bdgt.to_csv('../../Data/Estratos/Pathways/Contribution_miRNA_ConstantClusters_pathwayGO_BP.tsv', sep = '\t', index = False)
```
