```r
path <- '/home/ferambriz/Projects/SARS-CoV-2/Data/matrix_simetrized/'
files <- list.files(path)
files
```

```r
for (i in files){
    file = paste(path,i, sep ="")
    print(file)
    
    df <- read_tsv(file)
    x <- df$...1; df$...1 <- NULL;rownames(df) <-x
    df_norm <- as.data.frame(scale(df))

    s <- Spectrum(df_norm)
    df_norm$assignments <- s$assignments

    df_norm <- df_norm[,-c(1:dim(df_norm)[1] - 1)]
    print(df_norm)

    output = paste(path,i,'_clusters.tsv',sep ="")
    write.csv(df_norm, file=output)
}
```
