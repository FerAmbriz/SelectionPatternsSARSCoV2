```python
import pandas as pd
import os
import re
import itertools
```

```python
path = '/home/ferambriz/Projects/SARS-CoV-2/Data/matrix_simetrized/'
files = os.listdir(path)
files = [archivo for archivo in files if re.search(r"clusters\.tsv$", archivo)]
files
```

```python
lst = []
for i in files:
    df = pd.read_csv(f'{path}/{i}')
    lst_i = []
    for j in df['assignments'].unique():
        lst_i.append(list(df[df['assignments']==j]['Unnamed: 0']))
    lst.append(lst_i)
lst
```

```python
df_db = pd.DataFrame()

k = 0
for i in lst:
    df_db_i = pd.DataFrame()
    for j in i:
        comb_i = list(itertools.combinations(j, 2))
        df_i = pd.DataFrame(comb_i)
        df_db_i = pd.concat([df_db_i, df_i])
    df_db_i['Iteracion'] = k
    df_db = pd.concat([df_db, df_db_i])
    k+=1
df_db
#170
    
```

```python
lst_cm = []; lst_cnm = []

for i,j in zip(df_db[0], df_db[1]):
    if len(df_db[(df_db[0]==i)&(df_db[1]==j)]) == len(df_db['Iteracion'].unique()) and [i,j] not in lst_cm:
        lst_cm.append([i,j])
    else:
        lst_cnm.append([i,j])
lst_cm

#identificar que cluster son? o me creo unos nuevos?
```
