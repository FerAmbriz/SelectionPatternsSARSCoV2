```python
import pandas as pd
import os
import re
import itertools
import numpy as np
```

```python
path = '/home/ferambriz/Projects/SARS-CoV-2/Data/matrix_simetrized/'
files = os.listdir(path)
files = [archivo for archivo in files if re.search(r"clusters\.tsv$", archivo)]
files
```

```python
lst = []
for i in files:
    df = pd.read_csv(f'{path}/{i}')
    lst_i = []
    for j in df['assignments'].unique():
        lst_i.append(list(df[df['assignments']==j]['Unnamed: 0']))
    lst.append(lst_i)
lst
```

```python
df_db = pd.DataFrame()

k = 0
for i in lst:
    df_db_i = pd.DataFrame()
    for j in i:
        comb_i = list(itertools.combinations(j, 2))
        df_i = pd.DataFrame(comb_i)
        df_db_i = pd.concat([df_db_i, df_i])
    df_db_i['Iteracion'] = k
    df_db = pd.concat([df_db, df_db_i])
    k+=1
df_db
#170
    
```

```python
dicc = {}
for i,j in zip(df_db[0], df_db[1]):
    x = f'{i}-{j}'
    if x in dicc.keys():
        dicc[x] += 1
    else:
        dicc.update({x:1})
dicc
```

```python
def match_count_comparative(x,y):
    return dicc[f'{x}-{y}']

df_db['unique_comparison'] = list(map(match_count_comparative, df_db[0], df_db[1]))
df_db
```

```python
lst_cm = []; lst_cnm = []

for i,j in zip(df_db[0], df_db[1]):
    df_i = df_db[(df_db[0]==i) & (df_db[1]==j)]
    df_i = df_i.reset_index()
    print(df_i)
    # filtra los clusters que solo estan en 2, no esten en la lista unica e identifica los paises que estan maximos en su combinacion
    if (len(df_i) > 4 ) and ([i,j] not in lst_cm) and (len(df_i) == int(df_i['unique_comparison'][0])) :
        lst_cm.append([i,j])
    else:
        lst_cnm.append([i,j])
lst_cm

#identificar que cluster son? o me creo unos nuevos?
```

```python
uc = set()
for sublist in lst_cm:
    uc.update(sublist)

uc = list(uc)
uc
```

```python
lst_g = []
for i in uc:
    lst_gi = [i]
    for j in lst_cm:
        if i in j:
            jj = [element for element in j if element != i]
            lst_gi.append(jj[0])
    lst_g.append(set(lst_gi))
lst_g
```

```python
conjuntos_combinados = []

# Iterar a travÃ©s de los conjuntos
for conjunto in lst_g:
    # Verificar si hay intersecciÃ³n con algÃºn conjunto combinado previo
    combinado_previo = None
    for c in conjuntos_combinados:
        if conjunto & c:
            combinado_previo = c
            break

    # Si hay intersecciÃ³n, combinar los conjuntos
    if combinado_previo:
        combinado_previo |= conjunto
    else:
        # Si no hay intersecciÃ³n, agregar el conjunto como uno nuevo
        conjuntos_combinados.append(conjunto)

print("Conjuntos combinados:", conjuntos_combinados)
```
